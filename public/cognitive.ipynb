# OBSTACLE DETECTION USING ULTRASONIC SENSOR

```python
const int trigPin = 8;
const int echoPin = 10;
const int led_pin = 13;
float duration;
float distance;
void setup() {
    pinMode(led_pin, OUTPUT);
    pinMode(trigPin, OUTPUT);
    pinMode(echoPin, INPUT);
    Serial.begin(9600);
    Serial.println("Ultrasonic Sensor Active...");
}
void loop() {
    // Clear the trigPin
    digitalWrite(trigPin, LOW);
    delayMicroseconds(2);
    // Trigger the sensor by sending a 10 microsecond pulse
    digitalWrite(trigPin, HIGH);
    delayMicroseconds(10);
    digitalWrite(trigPin, LOW);
    // Read the echoPin, returns the sound wave travel time in microseconds
    duration = pulseIn(echoPin, HIGH);
    // Calculate the distance (Speed of sound is 0.034 cm/us)
    distance = duration * 0.034 / 2;
    // Print distance to Serial Monitor
    Serial.print("Distance: ");
    Serial.print(distance);
    Serial.println(" cm");
    // Logic for the 50cm threshold
    if (distance < 50) {
    digitalWrite(led_pin, HIGH);
    } else {
    digitalWrite(led_pin, LOW);
}
}
```

# Develop an embedded program to move the smart car in desired directions and control its velocity.

```python
// Pin configurations based on V5 expansion board mapping
#define ENA 6 // Right motor speed (PWM)
#define IN3 9 // Right motor direction 1
#define IN4 11 // Right motor direction 2
#define ENB 5 // Left motor speed (PWM)
#define IN1 7 // Left motor direction 1
#define IN2 8 // Left motor direction 2
void setup() {
// Set all motor control pins as outputs
    pinMode(ENA, OUTPUT);
    pinMode(IN3, OUTPUT);
    pinMode(IN4, OUTPUT);
    pinMode(ENB, OUTPUT);
    pinMode(IN1, OUTPUT);
    pinMode(IN2, OUTPUT);
}
void loop() {
// Movement Sequence
    moveForward(180); // Move forward at medium speed
    delay(2000);
    stopCar();

    delay(500);
    moveBackward(180); // Move backward
    delay(2000);
    pivotLeft(150); // Spin left in place
    delay(1000);
    stopCar();
    delay(2000);
}
// --- Movement Functions ---
// Controls both motors to move forward
void moveForward(int speed) {
    analogWrite(ENA, speed);
    analogWrite(ENB, speed);
    digitalWrite(IN1, HIGH); digitalWrite(IN2, LOW); // Left Forward
    digitalWrite(IN3, LOW); digitalWrite(IN4, HIGH); // Right Forward
}
// Controls both motors to move backward
void moveBackward(int speed) {
    analogWrite(ENA, speed);
    analogWrite(ENB, speed);
    digitalWrite(IN1, LOW); digitalWrite(IN2, HIGH); // Left Backward
    digitalWrite(IN3, HIGH); digitalWrite(IN4, LOW); // Right Backward
}
// Rotates the car to the left by reversing one wheel
void pivotLeft(int speed) {
    analogWrite(ENA, speed);
    analogWrite(ENB, speed);
    digitalWrite(IN1, LOW); digitalWrite(IN2, HIGH); // Left Backward
    digitalWrite(IN3, LOW); digitalWrite(IN4, HIGH); // Right Forward
}
// Rotates the car to the right
void pivotRight(int speed) {
    analogWrite(ENA, speed);
    analogWrite(ENB, speed);
    digitalWrite(IN1, HIGH); digitalWrite(IN2, LOW); // Left Forward
    digitalWrite(IN3, HIGH); digitalWrite(IN4, LOW); // Right Backward
}
// Stops all motor movement

void stopCar() {
    digitalWrite(IN1, LOW); digitalWrite(IN2, LOW);
    digitalWrite(IN3, LOW); digitalWrite(IN4, LOW);
    analogWrite(ENA, 0); analogWrite(ENB, 0);
}
```

# Develop an embedded program to move the smart car in desired directions and control its velocity.


```python
#define ENA 6 // Right motor speed

#define IN3 9 // Right motor direction 1
#define IN4 11 // Right motor direction 2
#define ENB 5 // Left motor speed
#define IN1 7 // Left motor direction 1
#define IN2 8 // Left motor direction 2
void setup() {
// Set all pins as outputs
    pinMode(ENA, OUTPUT);
    pinMode(IN3, OUTPUT);
    pinMode(IN4, OUTPUT);
    pinMode(ENB, OUTPUT);
    pinMode(IN1, OUTPUT);
    pinMode(IN2, OUTPUT);
    // Set speed (0-255)
    analogWrite(ENA, 170);
    analogWrite(ENB, 170);
}
void loop() {
    int speed = 120;
    for (int side = 1; side <= 4; side++) {
    analogWrite(ENA, speed);
    analogWrite(ENB, speed);
    moveBackward();
    delay(2000);
    if (side < 4) {
    turnLeft();
    delay(1000);
    }
    speed = speed + 45;
    if (speed > 255) {
    speed = 255;
    }
    }
    stopCar();
    delay(2000); // pause before restarting
}
// --- Movement Functions ---
void moveBackward() {
    digitalWrite(IN1, LOW); digitalWrite(IN2, HIGH); // Left backward
    digitalWrite(IN3, HIGH); digitalWrite(IN4, LOW); // Right backward
}

void turnLeft() {
    digitalWrite(IN1, LOW); digitalWrite(IN2, LOW); // Left stop
    digitalWrite(IN3, HIGH); digitalWrite(IN4, LOW); // Right backward
}
void turnRight() {
    digitalWrite(IN1, LOW); digitalWrite(IN2, HIGH); // Left backward
    digitalWrite(IN3, LOW); digitalWrite(IN4, LOW); // Right stop
}
void stopCar() {
    digitalWrite(IN1, LOW); digitalWrite(IN2, LOW);
    digitalWrite(IN3, LOW); digitalWrite(IN4, LOW);
}
```

# Develop an embedded program to implement obstacle avoidance using an ultrasonic sensor in the smart car.

```python
#include <Servo.h>
// ---------------- MOTOR PINS ----------------
#define ENA 6
#define IN3 9
#define IN4 11
#define ENB 5
#define IN1 7
#define IN2 8
// ---------------- ULTRASONIC + SERVO ----------------
#define ECHO A4
#define TRIG A5
#define SERVO_PIN 3
#define OBSTACLE_DISTANCE 20
Servo scanServo;
// ---------------------------------------------------
void setup() {
    pinMode(ENA, OUTPUT);
    pinMode(IN3, OUTPUT);
    pinMode(IN4, OUTPUT);
    pinMode(ENB, OUTPUT);
    pinMode(IN1, OUTPUT);
    pinMode(IN2, OUTPUT);
    pinMode(TRIG, OUTPUT);
    pinMode(ECHO, INPUT);
    scanServo.attach(SERVO_PIN);
    scanServo.write(90); // face forward
}
// ---------------------------------------------------
void loop() {
    long frontDist = getDistance();
    if (frontDist < OBSTACLE_DISTANCE) {

    avoidObstacle(); // obstacle has full control
    } else {
    moveForward(120); // free roaming forward
    }
}
// ---------------- MOVEMENT FUNCTIONS ----------------
void moveForward(int speed) {
    analogWrite(ENA, speed);
    analogWrite(ENB, speed);
    digitalWrite(IN1, HIGH); digitalWrite(IN2, LOW);
    digitalWrite(IN3, LOW); digitalWrite(IN4, HIGH);
}
void moveBackward(int speed) {
    analogWrite(ENA, speed);
    analogWrite(ENB, speed);
    digitalWrite(IN1, LOW); digitalWrite(IN2, HIGH);
    digitalWrite(IN3, HIGH); digitalWrite(IN4, LOW);
}
void pivotLeft(int speed) {
    analogWrite(ENA, speed);
    analogWrite(ENB, speed);
    digitalWrite(IN1, LOW); digitalWrite(IN2, HIGH);
    digitalWrite(IN3, LOW); digitalWrite(IN4, HIGH);
}
void pivotRight(int speed) {
    analogWrite(ENA, speed);
    analogWrite(ENB, speed);
    digitalWrite(IN1, HIGH); digitalWrite(IN2, LOW);
    digitalWrite(IN3, HIGH); digitalWrite(IN4, LOW);
}
void stopCar() {
    digitalWrite(IN1, LOW); digitalWrite(IN2, LOW);
    digitalWrite(IN3, LOW); digitalWrite(IN4, LOW);
    analogWrite(ENA, 0);
    analogWrite(ENB, 0);
}
// ---------------- DISTANCE FUNCTIONS ----------------

long getDistance() {
    digitalWrite(TRIG, LOW);
    delayMicroseconds(2);
    digitalWrite(TRIG, HIGH);
    delayMicroseconds(10);
    digitalWrite(TRIG, LOW);
    long duration = pulseIn(ECHO, HIGH, 25000);
    if (duration == 0) return 999;
    return duration * 0.034 / 2;
}
// ---------------- OBSTACLE AVOIDANCE ----------------
void avoidObstacle() {
    stopCar();
    delay(200);
    moveBackward(120);
    delay(400);
    stopCar();
    delay(150);
    scanServo.write(150);
    delay(300);
    long leftDist = getDistance();
    scanServo.write(30);
    delay(300);
    long rightDist = getDistance();
    scanServo.write(90);
    delay(200);
    if (leftDist > rightDist) {
    pivotLeft(150);
    delay(600);
    } else {
    pivotRight(150);
    delay(600);
    }
    stopCar();

    delay(150);
}
```

# Develop an embedded program to implement a line follower using infrared sensors in the smart car.

```python
// --- Motor Pins ---
#define ENA 6
#define ENB 5
#define IN1 7
#define IN2 8
#define IN3 9
#define IN4 11

// --- IR Sensor Pins ---
#define LEFT_IR 2
#define MID_IR 4
#define RIGHT_IR 10

// --- Settings ---
#define SPEED 100
#define TURN_SPEED 180

void setup() {
  // Motor pins
  pinMode(IN1, OUTPUT); 
  pinMode(IN2, OUTPUT);
  pinMode(IN3, OUTPUT); 
  pinMode(IN4, OUTPUT);
  pinMode(ENA, OUTPUT); 
  pinMode(ENB, OUTPUT);

  // IR sensors
  pinMode(LEFT_IR, INPUT);
  pinMode(MID_IR, INPUT);
  pinMode(RIGHT_IR, INPUT);

  Serial.begin(9600);
}

void loop() {

  int leftVal = digitalRead(LEFT_IR);
  int midVal = digitalRead(MID_IR);
  int rightVal = digitalRead(RIGHT_IR);

  // Debugging
  Serial.print("L:"); 
  Serial.print(leftVal);
  Serial.print(" M:"); 
  Serial.print(midVal);
  Serial.print(" R:"); 
  Serial.println(rightVal);

  // --- Line Following Logic ---
  if (midVal == LOW) {
    moveForward();   // middle sensor sees line → go straight
  }
  else if (leftVal == LOW) {
    turnLeft();      // left sensor sees line → adjust left
  }
  else if (rightVal == LOW) {
    turnRight();     // right sensor sees line → adjust right
  }
  else {
    stopMove();      // no sensor sees line → stop
  }
}

// --- Motor Control Functions ---

void moveForward() {
  analogWrite(ENA, SPEED);
  analogWrite(ENB, SPEED);

  digitalWrite(IN1, HIGH); 
  digitalWrite(IN2, LOW);   // Left motor forward

  digitalWrite(IN3, LOW);  
  digitalWrite(IN4, HIGH);  // Right motor forward (inverted logic)
}

void turnLeft() {
  analogWrite(ENA, TURN_SPEED);
  analogWrite(ENB, TURN_SPEED);

  digitalWrite(IN1, LOW);  
  digitalWrite(IN2, HIGH);  // Left motor backward

  digitalWrite(IN3, LOW);  
  digitalWrite(IN4, HIGH);  // Right motor forward
}

void turnRight() {
  analogWrite(ENA, TURN_SPEED);
  analogWrite(ENB, TURN_SPEED);

  digitalWrite(IN1, HIGH); 
  digitalWrite(IN2, LOW);   // Left motor forward

  digitalWrite(IN3, HIGH); 
  digitalWrite(IN4, LOW);   // Right motor backward
}

void stopMove() {
  analogWrite(ENA, 0);
  analogWrite(ENB, 0);

  digitalWrite(IN1, LOW); 
  digitalWrite(IN2, LOW);
  digitalWrite(IN3, LOW); 
  digitalWrite(IN4, LOW);
}
```

# Using a Raspberry Pi (RPi) and Python, write a program to control an LED connected to a GPIO pin such that:

A simple arithmetic function (for example, addition, multiplication, or any user-defined mathematical expression) is evaluated.
The numerical result of that arithmetic operation determines the number of times the LED flashes.
Each flash should consist of the LED turning ON for 1 second and OFF for 1 second.
You may use an appropriate Python GPIO library available on Raspberry Pi.

```python
import RPi.GPIO as GPIO
import time
import re

LED = 17

GPIO.setmode(GPIO.BCM)
GPIO.setup(LED, GPIO.OUT)

def calculate(expression):
    match = re.fullmatch(r"\s*(\d+)\s*([\+\-\*/])\s*(\d+)\s*", expression)

    if not match:
        return None

    num1 = int(match.group(1))
    operator = match.group(2)
    num2 = int(match.group(3))

    if operator == '+':
        return num1 + num2

    elif operator == '-':
        return num1 - num2

    elif operator == '*':
        return num1 * num2

    elif operator == '/':
        if num2 == 0:
            return None
        return int(num1 / num2)

try:
    while True:
        user_input = input("Enter calculation: ")
        result = calculate(user_input)

        if result is None or result < 0:
            continue

        for _ in range(result):
            GPIO.output(LED, GPIO.HIGH)
            time.sleep(0.3)
            GPIO.output(LED, GPIO.LOW)
            time.sleep(0.3)

except KeyboardInterrupt:
    GPIO.cleanup()
```

# Write a Python program to capture 1) a picture and 2) a series of 5 pictures every 10 seconds using RPi and the RPi camera.


```python
from picamera2 import Picamera2
import time
from datetime import datetime
import cv2

picam2 = Picamera2()

config = picam2.create_still_configuration()
picam2.configure(config)

picam2.start()
time.sleep(2)

frame = picam2.capture_array()

frame_bgr = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

height, width, _ = frame_bgr.shape

font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
color = (0, 255, 0)
thickness = 2

positions = {
    "Aakarshak": (50, height - 50),
    "Abhilash": (width // 2 - 50, height - 50),
    "Shashanth": (width - 250, height - 50)
}

for name, pos in positions.items():
    cv2.putText(
        frame_bgr,
        name,
        pos,
        font,
        font_scale,
        color,
        thickness,
        cv2.LINE_AA
    )

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"annotated_image_{timestamp}.jpg"

cv2.imwrite(filename, frame_bgr)

print(f"Image captured and saved as {filename}")

picam2.stop()
```

```python
from picamera2 import Picamera2
import time
from datetime import datetime
import os

num_captures = 5
interval_seconds = 5
save_dir = "captures"

os.makedirs(save_dir, exist_ok=True)

picam2 = Picamera2()
config = picam2.create_still_configuration()
picam2.configure(config)

picam2.start()
time.sleep(2)

print(f"Starting timed capture: {num_captures} images every {interval_seconds}s")

start_time = time.time()

for i in range(1, num_captures + 1):
    target = start_time + (i - 1) * interval_seconds

    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = os.path.join(save_dir, f"image_{timestamp}_{i:02d}.jpg")

    picam2.capture_file(filename)
    print(f"[{i}/{num_captures}] Saved {filename}")

    if i < num_captures:
        remaining = target + interval_seconds - time.time()
        if remaining > 0:
            time.sleep(remaining)

picam2.stop()
print("Done.")
```
